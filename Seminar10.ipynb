1. Изобразите отношение households к population с помощью точечного графика
2. Визуализировать longitude по отношения к median_house_value, используя линейный график
3. Представить гистограмму по housing_median_age
Изобразить гистограмму по median_house_value с оттенком housing_median_age

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
df.sample(5)
```

```python
plt.figure(figsize = (8,4))
sns.scatterplot(data=df,x="households",y="population",hue="total_rooms",size=10)
```

```python
cols = ['population','median_income','housing_median_age','median_house_value']
g = sns.PairGrid(df[cols])
g.map(sns.scatterplot)
```

2. Визуализировать longitude по отношения к median_house_value, используя линейный график

```python
sns.relplot(x = 'longitude',y = 'median_house_value',data = df, aspect = 1.5, height = 3)
sns.relplot(x = 'longitude',y = 'median_house_value',kind = 'line',data = df, aspect = 1.5, height = 3)
```

```python
plt.figure(figsize = (8,4))
sns.lineplot(x = 'longitude',y = 'median_house_value',data = df)
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(data = df, x = 'longitude', y = 'latitude', hue = 'median_house_value')
```

3. Представить гистограмму по housing_median_age Изобразить гистограмму по median_house_value с оттенком housing_median_age

```python
plt.figure(figsize = (8,4))
plt.title('Динамика housing_median_age')
sns.histplot(data = df, x = 'housing_median_age')
```

```python
plt.figure(figsize = (8,4))
plt.title('median_house_value')
sns.histplot(data = df[df['housing_median_age'] > 50], x = 'median_house_value')
```

```python
df.locc[df['housing_median_age'] <= 20, 'age_group'] = 'Молодые'
df.loc[(df['housing_median_age'] > 20) & (df['housing_median_age'] <= 50),
'age_group'] = 'Ср. возраст'
df.loc[df['housing_median_age'] > 50, 'age_group'] = 'Пожилые'
```
